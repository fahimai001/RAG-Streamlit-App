```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("../data/attention_is_all_you_need.pdf")
```

```python
documents_loader = loader.load()
```

```python
documents_loader
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 500)

documents = splitter.split_documents(documents_loader)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(google_api_key=api_key, model="models/embedding-001")
```

```python
from dotenv import load_dotenv
load_dotenv()
weaviate_api_key = os.getenv("WEAVIATE_API_KEY")
weaviate_url = os.getenv("WEAVIATE_URL")
```

```python
import weaviate
from weaviate.classes.init import Auth

client = weaviate.connect_to_weaviate_cloud(
    cluster_url=weaviate_url,
    auth_credentials=Auth.api_key(weaviate_api_key),
)
```

```python
from langchain_weaviate.vectorstores import WeaviateVectorStore


vector_store = WeaviateVectorStore.from_documents(documents, embeddings, client=client)
```
